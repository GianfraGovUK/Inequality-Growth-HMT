```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"data\growth_data.csv")
```

```python
data
```

```python
data["lnY"] = np.log(data["GDP_pc"])
data["lnGINI"] = np.log(data["gini"])
data["lnINF"] = np.log(data["change_in_cpi"])
data["lnW"] = np.log(data["wrl_gdp_growth"])
```

```python
data
```

```python
data = data.dropna()
```

```python
import statsmodels.api as sm

# Prepare the feature matrix with two separate constants
X = data[['lnGINI', 'lnINF', "lnW"]].copy()

# Define the dependent variable
y = data['lnY']

# Fit the model and print results
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from statsmodels.tsa.ardl import ARDL

# Example ARDL with one lag for each regressor
ardl_model = ARDL(endog=y, exog=X[['lnGINI', 'lnINF', "lnW"]], lags=3, order={'lnGINI': 1, 'lnINF': 3, 'lnW': 0})
ardl_result = ardl_model.fit()
print(ardl_result.summary())
```
